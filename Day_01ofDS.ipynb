```python
lst = []
lst1 =[]
lst2 = []
choice = input("type 'y' if you want to continue , 'n' to exit !!!") 
while choice == 'y':
    batsman_name = input("ENTER A BATSMAN NAME :")
    scores = int(input("ENTER THE RUNS OF THE BATSMAN :"))
    lst1.append([batsman_name,scores])
    bowler_name = input("ENTERA BOWLER'S NAME :")
    wickets=int(input("ENTER THE WICKETS of this bowler : "))
    lst2.append([bowler_name,wickets])

    choice = input("type 'y' if you want to continue , 'n' to exit !!!")

lst.append([lst1,lst2])
print(lst)
```

```python

price=0.0
choice=input("TYPE Y TO CONTINUE N TO EXIT")
while choice == 'y':
    unit = float(input("ENTER YOU UNIT OF ELECTRICITY BILL :"))
    if unit<=200.0 :
        print("TOTAL PRICE : ",price)
    elif 200.0<unit and unit<= 300.0:
        price=(unit-200)*5
        print("TOTAL PRICE :", price)
    elif unit>300.0:
        price=(unit-300)*10 + (100*5)
        print("TOTAL PRICE",price)
    else :
        print("INVALID CHOICE !!!")
    choice=input("TYPE Y TO CONTINUE N TO EXIT")
```
