## Q1. Student Information System
### Create a Student class with attributes: Name, Roll Number, Marks (3 subjects).
### Methods: get_data(), display_data(), calculate_percentage().

```python
class Student :
    def get_data(self):
        name = input("ENTER YOUR NAME :")
        roll_number = int(input("ENTER YOUR ROLL NUMBER :"))
        CN = float(input("ENTER MARKS OF CN :"))
        OS = float(input("ENTER MARKS OF OS :"))
        OOPS = float(input("ENTER MARKS OF OOPS :"))
        self.name=name
        self.roll_number=roll_number
        self.CN=CN
        self.OS=OS
        self.OOPS=OOPS
    def display_data(self):
        choice = int(input('''1.Display name
2.Display roll number
3.Display CN marks
4.Display OS marks
5.Display OOPS marks
'''))
        if choice == 1:
            print("YOUR NAME IS :",self.name)
        elif choice == 2:
            print("YOUR ROLL NUMBER IS :",self.roll_number)
        elif choice == 3:
            print("Marks of CN :",self.CN)
        elif choice == 4:
            print("Marks of OS :",self.OS)
        elif choice == 5:
            print("Marks of OOPS :",self.OOPS)
            
    def calculate_percentage(self):
        print("YOUR PERCENTAGE IS :",((self.CN + self.OS + self.OOPS)/300)*100)
s = Student()
while True :
    choice = int(input('''1-add student
2-show student datils
3.calculate percentage
4.exit'''))
    if choice == 1:
        s.get_data()
        print("DETAILS ADDED !!!")
    elif choice == 2:
        s.display_data()
    elif choice == 3:
        s.calculate_percentage()
    elif choice == 4:
        break
```

## Q2. Bank Account
### Create a BankAccount class with Account Holder Name, Account Number, Balance.
### Methods: deposit(), withdraw(), display_balance().
### Conditions: Deposit must be positive and withdrawal cannot exceed balance.


```python
class account :
    def __init__(self,name="",acc_no=0,bal=None):
        if bal is None :
            bal = [0]
        self.name=name
        self.acc_no=acc_no
        self.bal=bal
    def deposit(self):
        num=int(input("ENTER AMOUNT TO ADD:"))
        self.bal = [num]
    def withdraw(self):
        num1=int(input("ENTER AMOUNT TO WITHDRAW:"))
        self.bal = [num1]
    def display_balance(self):
        print(self.bal)
a = account()
a.display_balance()
a.deposit()
a.display_balance()
a.withdraw()
a.display_balance()
    
```

## Q3. Employee Salary Calculator
### Create an Employee class with Employee ID, Name, Basic Salary.
### Methods to calculate HRA (20%), DA (10%), Gross Salary, and display details.


```python
class emp:
    def __init__(self,name="",emp_id=0,sal=None):
        if sal is None:
            sal = [0]
        self.name=name
        self.emp_id=emp_id
        self.sal=sal
    def get(self):
        self.name=input("ENTER YOUR NAME:")
        self.emp_id=int(input("ENTER YOUR EMPLOYEE ID:"))
        self.sal=float(input("Enter your basic salary:"))
    def HRA(self):
        self.HRA=self.sal*0.2
    def DA(self):
        self.DA=self.sal*0.1
    def GS(self):
        self.GS=self.sal+self.HRA+self.DA
    def display(self):
        print("NAME:",self.name)
        print("EMPLOYEE ID:",self.emp_id)
        print("BASIC SALARY=:",self.sal)
        print("HRA:",self.HRA)
        print("DA:",self.DA)
        print("GS:",self.GS)

e=emp()
e.get()
e.HRA()
e.DA()
e.GS()
e.display()
```

## Q4. Rectangle Operations
### Create a Rectangle class with Length and Breadth.
### Methods: area(), perimeter(), check_square().


```python
class rectangle:
    def base(self):
        l = int(input("ENTER THE LENGTH :"))
        b = int(input("ENTER THE BREADTH :"))
        self.l=l
        self.b=b
    def area(self):
        self.area=self.l*self.b
        print("AREA :",self.area,"Sq. unit")
    def per(self):
        self.per = 2*(self.l + self.b)
        print("PERIMETER ;",self.per,"unit")
    def check_square(self):
        if self.l == self.b :
            print("IT'S A SQUARE!!!")
        else :
            print("IT'S A RECTANGLE!!!")
r = rectangle()
r.base()
r.area()
r.per()
r.check_square()
```
